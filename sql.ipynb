```python
import csv, sqlite3
import prettytable
import pandas as pd
prettytable.DEFAULT = 'DEFAULT'
con = sqlite3.connect('walmart_sales.db')
cur = con.cursor()
%sql sqlite:///walmart_sales.db
df = pd.read_csv('walmart_sales.csv')
df.to_sql('WALMART_SALES_TABLE', con, if_exists='replace', index=False)
```

```python
pd.read_sql('SELECT * FROM WALMART_SALES_TABLE WHERE Weekly_Sales < (SELECT AVG(Weekly_Sales) FROM WALMART_SALES_TABLE) ORDER BY Date DESC', con)
```

```python
pd.read_sql('SELECT Date, Weekly_Sales, Temperature FROM WALMART_SALES_TABLE WHERE Holiday_Flag = 0 ORDER BY Weekly_Sales DESC LIMIT 10', con)
```

```python
pd.read_sql('SELECT Temperature, Weekly_Sales FROM WALMART_SALES_TABLE ORDER BY Temperature DESC', con)
```
